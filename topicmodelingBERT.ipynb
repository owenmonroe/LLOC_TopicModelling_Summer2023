```python
def preprocess_data(clean_text_df):
   # speech_string = ''.join(clean_text_df)
    clean_text = remove_stopwords(clean_text_df)
    lemmatizer = WordNetLemmatizer()
    l_text = lemmatizer.lemmatize(clean_text, pos='n')
    no_punc = l_text.replace(',', "").replace('.', "").replace('(', "").replace(')', "").replace('/', "").replace('\\n', "").replace(':', "").replace(';', "").replace('-', " ").replace('?', "")
    final_text_df = no_punc.lower()
    
    return final_text_df
```
